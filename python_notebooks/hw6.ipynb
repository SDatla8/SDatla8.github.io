```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
data_url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv"
df = pd.read_csv(data_url)

df.loc[df['Year Constructed'] == 0, 'Year Constructed'] = np.nan
df.loc[df['Square Footage'] == 0, 'Square Footage'] = np.nan
```

```python
selection = alt.selection_point(fields=['Bldg Status'], bind='legend')

chart1 = alt.Chart(data_url).mark_bar().encode(
    alt.Y('Agency Name:N', sort='-x'),
    alt.X('count():Q'),
    alt.Color('Bldg Status:N'),
    opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
).add_params(
    selection
).properties(
    width=400,
    height=500
)

chart1
```

```python
chart2 = alt.Chart(data_url).mark_rect().encode(
    alt.X('Usage Description:N'),
    alt.Y('Bldg Status:N'),
    alt.Color('count():Q', scale=alt.Scale(scheme='purples')),
    tooltip=[
        alt.Tooltip('Usage Description:N'), 
        alt.Tooltip('Bldg Status:N'), 
        alt.Tooltip('count():Q')
    ]
).properties(
    width=500,
    height=200
).interactive()

chart2
```
